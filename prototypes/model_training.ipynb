```python
import pandas as pd

import auto_ml
from sklearn.model_selection import train_test_split
from auto_ml import Predictor
from auto_ml.utils import get_boston_dataset
```

```python
df = pd.read_parquet('sale_final_2019_11_04T20_20_49.parquet')
print(df.shape)
df = df.drop_duplicates(subset=['lat'], keep='last')
print(df.shape)
```

```python
df['coords_factor'] = df['coords_mean_price_m2'] + (df['coords_mean_price_m2'] / (df['coords_cluster_center_dist_km'] + 1))
df.head()
```

```python
for c in df.columns:
    print(c)
```

```python
df = df.drop(['price', 'offer_id', 'date_refreshed'], axis=1)

df_train, df_test = train_test_split(df, test_size=0.2)

column_descriptions = {
    'price_m2': 'output',
    'date_added': 'date',
    'coords_category': 'categorical',
    'balcony': 'categorical',
    'building_material': 'categorical',
    'building_type': 'categorical',
    'heating': 'categorical',
}

ml_predictor = Predictor(type_of_estimator='regressor', column_descriptions=column_descriptions)

ml_predictor.train(df_train)

ml_predictor.score(df_test, df_test['price_m2'])
```
